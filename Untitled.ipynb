```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Đọc file

```python
df = pd.read_csv("household_power_consumption.txt", delimiter=";", low_memory = False)
```

```python
df
```

### Kiểm tra kiểu dữ liệu

```python
df.dtypes
```

### Tiền xử lí

#### Đổi các dấu '?' thành nan

```python
df[['Global_active_power','Global_reactive_power','Voltage','Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']] = df[['Global_active_power','Global_reactive_power','Voltage','Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']].replace('?', np.nan)
```

#### Đổi sang kiểu dữ liệu số

```python
df[['Global_active_power','Global_reactive_power','Voltage','Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']] = df[['Global_active_power','Global_reactive_power','Voltage','Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']].astype('float64')
```

#### Đổi sang kiểu dữ liệu ngày

```python
df['Date'] =  pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Time'] = pd.to_timedelta(df['Time'])
```

#### Ghép Date và Time vào cột mới là 'Period'

```python
df['Period'] = pd.to_datetime(df['Date'] + df['Time'], format='%Y-%m-%d%H:%M:%S')
```

#### Lọc ngày '2007-02-01' và '2007-02-02'

```python
df = df[df["Date"].isin(pd.date_range('2007-02-01', '2007-02-02'))]
```

```python
df
```

### Plot 1

```python
def plot1():
    plt.hist(df.Global_active_power, bins = np.arange(0, 7, 0.5), color ='r', edgecolor = "black")
    plt.xlabel('Global Active Power (kilowatts)') 
    plt.ylabel('Frequency')
    plt.title('Global Active Power')
    plt.savefig('figure1.png')
    plt.show()

plot1()
```

### Plot 2

```python
def plot2():
    plt.figure(figsize=(15, 10))
    
    plt.plot(df.Period, df.Global_active_power, color = 'black')
    plt.ylabel('Global Active Power (kilowatts)')
    plt.savefig('figure2.png')
    plt.show()

plot2()
```

### Plot 3

```python
def plot3():
    plt.figure(figsize=(10, 5))

    plt.plot(df.Period, df.Sub_metering_1, color = 'gray', label = 'Sub_metering_1')
    plt.plot(df.Period, df.Sub_metering_2, color = 'red', label = 'Sub_metering_2')
    plt.plot(df.Period, df.Sub_metering_3, color = 'blue', label = 'Sub_metering_3')
    plt.legend()
    plt.ylabel('Energy sub metering')
    plt.savefig('figure3.png')
    plt.show()

plot3()
```

### Plot 4

```python
def plot4():

    figure, axis = plt.subplots(2, 2)
    axis[0, 0].plot(df.Period, df.Global_active_power, color = 'black')
    axis[0, 0].set_ylabel('Global Active Power (kilowatts)')
    
    axis[0, 1].plot(df.Period, df.Voltage, color = 'black')

    axis[1, 0].plot(df.Period, df.Sub_metering_1, color = 'gray', label = 'Sub_metering_1')
    axis[1, 0].plot(df.Period, df.Sub_metering_2, color = 'red', label = 'Sub_metering_2')
    axis[1, 0].plot(df.Period, df.Sub_metering_3, color = 'blue', label = 'Sub_metering_3')
    axis[1, 0].legend()

    axis[1, 1].plot(df.Period, df.Global_reactive_power, color = 'black')
    f = plt.gcf()
    f.set_figwidth(25)  
    f.set_figheight(15)
    plt.show()

plot4()
```
